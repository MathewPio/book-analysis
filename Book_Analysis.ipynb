```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

```python
type(book)
```

```python
# regex
```

```python
import re
```

```python
pattern = re.compile("Chapter [0-9]+")
findings = re.findall(pattern, book)
len(findings)
```

# Which are the sentences love was used?

```python
pattern = re.compile("[^.]* love [^.]*.")
findings = re.findall(pattern, book)
findings
```
